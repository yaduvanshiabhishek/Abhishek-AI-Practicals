```python
# Step 1: Import Libraries
import numpy as np
import matplotlib.pyplot as plt

# Step 2: Data Collection
X = np.array([10, 20, 30, 40, 50, 60, 70, 80])   # Advertising Cost
Y = np.array([100, 150, 200, 250, 300, 350, 400, 450])  # Product Sales

# Step 3: Calculation (slope m and intercept c)
n = len(X)

sum_x = np.sum(X)
sum_y = np.sum(Y)
sum_xy = np.sum(X * Y)
sum_x2 = np.sum(X * X)

# Slope (m)
m = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - (sum_x ** 2))

# Intercept (c)
c = (sum_y - m * sum_x) / n

print("Slope (m):", m)
print("Intercept (c):", c)

# Step 4: Prediction (User Input)
x_new = float(input("Enter Advertising Cost to predict Sales: "))
y_pred_new = m * x_new + c

print("Predicted Sales for", x_new, "is:", y_pred_new)

# Step 5: Visualization
Y_pred = m * X + c

plt.scatter(X, Y)        # Actual data
plt.plot(X, Y_pred)      # Regression line
plt.xlabel("Advertising Cost")
plt.ylabel("Product Sales")
plt.title("Linear Regression - Advertising vs Sales")
plt.show()
```
